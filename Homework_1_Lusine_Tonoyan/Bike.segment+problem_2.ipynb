```python
import pandas as pd
```

```python
data=pd.read_excel("bike.xlsx",sheetname=3)
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
cluster=KMeans(5)
```

```python
inputs=data.iloc[:,1:]
```

```python
inputs.head()
```

```python
data["cluster"]=cluster.fit_predict(inputs)
```

```python
data.head()
```

```python
data["cluster"].value_counts()
```

```python
new_data=data.iloc[:,[0,-1]]
```

```python
new_data.head()
```

```python
new_data.to_excel("bike_cluster.xlsx")
```

```python
import pandas as pd
from statsmodels.formula.api import ols
```

```python
data = pd.read_excel("conjoint_survey.xlsx")
```

```python
data.head()
```

```python
st=pd.DataFrame(columns=range(33))
```

```python
for i in range(33):
    model=ols(formula="Rank ~ A1 + A2 + A3 + B1 + B2 + C1 + C2 + C3", data=data.iloc[i*12:i*12+12,1:])
    results = model.fit()
    coef = results.params
    st[i]=coef
```

```python
st.head()
```

```python
st.to_excel("coefficients.xlsx")
```
